```python
from bs4 import BeautifulSoup
import glob
```

```python
save_loc = "libguides*" 
```

```python
files = glob.glob(save_loc+"/*")
```

```python
i = 0
j = 0
num_files = 40
text = [""]
for file in files:
    print(i, j)
    if ".html" in file:
        with open(file, 'rb') as page:
            soup = BeautifulSoup(page.read())
        text[j] = text[j] + soup.text
        page.close()
    i = i + 1
    if not(i % num_files):
        j = j + 1
        text.append("")
    
```

```python
i = 1
for output in text:
    print(i)
    with open(str(i)+"_batch.txt", "w", encoding = "utf-8") as f:
        f.write(output)
    f.close()
    i = i + 1
```
